## 04. Decomposition Time series

### Additive Model for Time Series Analysis
If yt is the time series value at time t. Tt, St, Ct, and Rt are the trend value, seasonal, cyclic and random fluctuations at time t respectively. According to the Additive Model, a time series can be expressed as

                                        yt = Tt + St + Ct + Rt.

This model assumes that all four components of the time series act independently of each other.

### Multiplicative Model for Time Series Analysis
The multiplicative model assumes that the various components in a time series operate proportionately to each other. According to this model

                                        yt = Tt × St × Ct × Rt

```python
from pandas import read_csv
from matplotlib import pyplot as plt
series = read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv', header=0, index_col=0)
series.plot()
```

#### Reviewing the line plot, it suggests that there may be a linear trend, but it is hard to be sure from eye-balling. There is also seasonality, but the amplitude (height) of the cycles appears to be increasing, suggesting that it is multiplicative.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(series[:10], model='multiplicative',period=1)
result.plot()
pyplot.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(series[:10], model='additive',period=1)
result.plot()
pyplot.show()
```
